```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

```python

x,y = load_diabetes(return_X_y=True)
```

```python

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import SGDRegressor
reg = SGDRegressor(penalty='l2',max_iter=500,eta0=0.1,learning_rate='constant',alpha=0.001)
```

```python
reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=0.005, max_iter=500,solver='sparse_cg')
reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)
print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)
```

```python
class MeraRidgeGD:
    def __init__(self,alpha,epochs,lrt,):
        self.alpha = alpha
        self.epochs=epochs
        self.lrt=lrt
        self.coef_=None
        self.intercept_=None
    def fit(self,x,y):
        self.coef_=np.ones(x.shape[1])
        self.intercept_=0
        theta=np.insert(self.coef_,0,self.intercept_)
        x=np.insert(x,0,1,axis=1)
        for i in range(self.epochs):
            theta_der=np.dot(x.T,x).dot(theta)-np.dot(x.T,y)+self.alpha*theta
            theta=theta-self.lrt*theta_der
        self.coef_=theta[1:]
        self.intercept_=theta[0]    
        print(self.coef_,self.intercept_)
    def predict(self,x_test):
        return np.dot(x_test,self.coef_)+self.intercept_
```

```python
reg = MeraRidgeGD(epochs=500,alpha=0.001,lrt=0.005)
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
print("R2 score",r2_score(y_test,y_pred))

```
